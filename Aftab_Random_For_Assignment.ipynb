# **Splitting Data**

```python
from sklearn.model_selection import train_test_split
```

```python
x=df_scaled.drop('Type',axis=1)
y=df['Type']
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

# **Random Forest**

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier(n_estimators=100,random_state=0)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

```python
print('Accuracy:',accuracy_score(y_test,y_pred))
print('Precision:',precision_score(y_test,y_pred,average='macro'))
print('Recall:',recall_score(y_test,y_pred,average='macro'))
```

# **Bagging**

```python
from sklearn.ensemble import BaggingClassifier, AdaBoostClassifier
```

```python
bagging=BaggingClassifier(random_state=40)
```

```python
bagging.fit(x_train,y_train)
```

```python
y_pred_bagging=bagging.predict(x_test)
```

```python
print('Accuracy:',accuracy_score(y_test,y_pred_bagging))
```

# **Boosting**

```python
boosting=AdaBoostClassifier(random_state=25)
```

```python
boosting.fit(x_train,y_train)
```

```python
y_pred_boosting=boosting.predict(x_test)
```

```python
print('Accuracy:',accuracy_score(y_test,y_pred_boosting))
```

# **Interview Questions**

```python
'''
1.  Bagging is an ensemble learning method that:

       Creates multiple subsets of the training data (with replacement)
       Trains a model on each subset
       Combines the predictions of all models


    Boosting is an ensemble learning method that:

       Trains a model on the entire training data
       Calculates the errors of the model
       Trains another model on the errors
       Combines the predictions of both models

    Bagging is a parallel approach that reduces overfitting by averaging multiple models,
    while Boosting is a sequential approach that improves performance by focusing on errors and weighting predictions.'''
```
