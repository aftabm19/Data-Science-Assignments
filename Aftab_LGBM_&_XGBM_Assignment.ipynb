```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy='most_frequent')
training[['Age', 'Cabin','Embarked']] = imputer.fit_transform(training[['Age', 'Cabin','Embarked']])
```

```python
training.isnull().sum()
```

```python
training.drop('Name',axis=1,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
training['Sex']=le.fit_transform(training['Sex'])
training['Age']=le.fit_transform(training['Age'])
training['Embarked']=le.fit_transform(training['Embarked'])
training['Cabin']=le.fit_transform(training['Cabin'])
training['Ticket']=le.fit_transform(training['Ticket'])
```

```python
x=training.drop('Survived',axis=1)
y=training['Survived']
```

```python
x
```

```python
y
```

# **Splitting Data**

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

# **LGBM & XGBM models**

```python
import lightgbm as lgb
import xgboost as xgb
```

```python
lgbm=lgb.LGBMClassifier()
lgbm.fit(x_train,y_train)
```

```python
y_pred_lgb=lgbm.predict(x_train)
```

# **Evaluation**

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
print(y_test.shape)
print(y_pred_lgb.shape)
```

```python
y_pred_lgb = y_pred_lgb[:len(y_test)]         #it need to do same length
```

```python
y_test = y_test[:len(y_pred_lgb)]
```

```python
print(y_test.shape)
print(y_pred_lgb.shape)
```

```python
print("LightGBM:")
print("Accuracy:", accuracy_score(y_test, y_pred_lgb))
print("Precision:", precision_score(y_test, y_pred_lgb))
print("Recall:", recall_score(y_test, y_pred_lgb))
print("F1-score:", f1_score(y_test, y_pred_lgb))
```

```python
xgbm=xgb.XGBClassifier()
xgbm.fit(x_train,y_train)
```

```python
y_pred_xgb=xgbm.predict(x_test)
```

```python
print(y_test.shape)
print(y_pred_xgb.shape)
```

```python
print("XGBoost:")
print("Accuracy:", accuracy_score(y_test, y_pred_xgb))
print("Precision:", precision_score(y_test, y_pred_xgb))
print("Recall:", recall_score(y_test, y_pred_xgb))
print("F1-score:", f1_score(y_test, y_pred_xgb))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# LightGBM
cm_lgbm = confusion_matrix(y_test, y_pred_lgb)
ConfusionMatrixDisplay(cm_lgbm).plot()
plt.title("LightGBM Confusion Matrix")
plt.show()

# XGBoost
cm_xgb = confusion_matrix(y_test, y_pred_xgb)
ConfusionMatrixDisplay(cm_xgb).plot()
plt.title("XGBoost Confusion Matrix")
plt.show()
```
