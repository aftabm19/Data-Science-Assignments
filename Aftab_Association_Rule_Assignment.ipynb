```python
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
#!pip install mlxtend
```

```python
aray_data=data.to_numpy()
```

```python
aray_data
```

```python
te=TransactionEncoder()
```

```python
en_data=te.fit_transform(aray_data)
```

```python
en_data
```

```python
online_retail=pd.DataFrame(en_data,columns=te.columns_)
```

```python
online_retail
```

```python
freq_itemset = apriori(online_retail,min_support=0.1,use_colnames=True)
```

```python
freq_itemset
```

```python
rules = association_rules(freq_itemset,metric='lift',min_threshold=1)
rules.sort_values('lift',ascending=False).head(10)
```

```python
freq_itemset=apriori(online_retail,min_support=0.03,use_colnames=True)
rules_1=association_rules(freq_itemset,metric='lift',min_threshold=1)
rules_1.tail(10)
```

```python
plt.bar(rules['antecedents'].apply(lambda x:list(x)[0]),rules['confidence'])
plt.xlabel('Anticidents')
plt.ylabel("Confidence")
plt.title('Confidence Chart')
```

```python
'''
1. Lift is a metric which is used in association rule mining measures how much more likely
   two itemns are buying together and or not.
   The true association
   lift is greater than 1 then its indicate positive association, and items more likely purches together.

   lift is 1 then indicates no association, items are independent for purchase

   lift in smaller than 1 then its shows negative association, and items are less likey purchase together.
'''
```

```python
'''
2.Support and Confidence are the metrics of Association rules. They are used to evaluate the strength
  and usefullness for association analysis.

  SUPPORT
  Support counts how many time an itemset appears in the dataset.
  Formula of support is support = number of transaction of containing both A & B/ total number of transactions.

  COFIDENCE
  Confidence says how many times the rule has been found to be true,
  i.e., how many times B iten is purchased when A item is purchased.
  formula of confidence is: confidence A to B = number of transaction of containing both A & B/number of transactios containing A
'''
```
