```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df['type']=le.fit_transform(df['type'])
```

```python
df['type']
```

```python
df=df.replace('Unknown',0)
```

```python
x=df.iloc[:,[0,4,5,6]].astype(float)
y=df['type'].astype(float)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.15,random_state=52)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log=LogisticRegression()
```

```python
log.fit(train_x,train_y)
```

```python
model=log.fit(train_x,train_y)
```

```python
from sklearn.metrics import classification_report
```

```python
print('report for train data')
print(classification_report(train_y,model.predict(train_x)))
```

```python
precision_score(train_y,model.predict(train_x),average='weighted')
```

```python
print('report for test data')
print(classification_report(test_y,model.predict(test_x)))
```

```python
precision_score(test_y,model.predict(test_x),average='weighted')
```
