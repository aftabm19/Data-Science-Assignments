```python
df=df.drop('animal name',axis=1) # Dropping 'animal name' feature it's categorical & non-useful for classificarion.
```

```python
df
```

# **Scalling Data**

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
x=df.drop('type',axis=1) # Spltting into x and y
y=df['type']
```

```python
x
```

```python
y
```

```python
x_scaled=scaler.fit_transform(x)
```

```python
x_scaled
```

# **Splitting Into Train-Test**

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2,random_state=0)
```

# **KNN**

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
k=5
knn = KNeighborsClassifier(n_neighbors=k, metric = 'euclidean')
```

```python
knn.fit(x_train,y_train)
```

```python
y_pred = knn.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score,classification_report, confusion_matrix
```

```python
classification_report(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
```

```python
x_pca=pca.fit_transform(x_scaled)
```

```python
x_train_pca, x_test_pca, _, _ = train_test_split(x_pca, y, test_size=0.2, random_state=42)
```

```python
knn_pca = KNeighborsClassifier(n_neighbors=k)
```

```python
knn_pca.fit(x_train_pca,y_train)
```

```python
h=0.2
```

```python
x_min, x_max = x_pca[:, 0].min() - 1, x_pca[:, 0].max() + 1
y_min, y_max = x_pca[:, 1].min() - 1, x_pca[:, 1].max() + 1
```

```python
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))
```

```python
Z = knn_pca.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
```

```python
Z = knn_pca.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
```

```python
from matplotlib.colors import ListedColormap
```

```python

plt.figure(figsize=(10, 6))
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF', '#FFFACD', '#E0FFFF', '#FFC0CB', '#D3D3D3'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF', '#FFD700', '#00CED1', '#FF69B4', '#A9A9A9'])

plt.contourf(xx, yy, Z, cmap=cmap_light)
sns.scatterplot(x=x_pca[:, 0], y=x_pca[:, 1], hue=y, palette=cmap_bold, edgecolor='k')
plt.title("KNN Decision Boundary (2D PCA Projection)")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.show()
```
